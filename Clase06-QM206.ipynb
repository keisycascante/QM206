```python
A=1.0
B=0.0
k1=0.01
t=0
dt=0.1
cA=[]
cB=[]
tt=[]
while t<100:
  cA.append(A)
  cB.append(B)
  tt.append(t)
  dA=-k1*A*dt
  dB=k1*A*dt
  A=A+dA
  B=B+dB
  t=t+dt
plt.plot(tt,cA,label="[A]")
plt.plot(tt,cB,label="[B]")
plt.legend()
plt.show()
```

ke=(x)/(1.00-x)*(0.5-2x)^2=1x10^-3
f(x)=1E-3      

```python
def f(x):
  return (x)/((1.00-x)*(0.5-2*x)**2)
```

```python
ke=1E-3
x=0.0005
x=np.linspace(0.0002, 0.004, 1000)
v=(f(x)-ke)
plt.plot(x,v)
plt.show()
```
